Converting the numerical values into numeric

Cancer Data


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_data = pd.read_csv("/Breast Cancer dataset.csv")
```

```python
cancer_data.head()
```

```python
cancer_data.shape
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
cancer_data['Targets'] = labels
```

```python
cancer_data.head()
```

```python
cancer_data['Targets'].value_counts()
```
